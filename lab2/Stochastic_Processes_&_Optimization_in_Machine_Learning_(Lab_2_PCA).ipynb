<p>Then, you will compute the mean of each dataset column (feature) and normalize each feature using it</p>

```
M = mean(data.T, axis=1)
data_normal = data - M
```

<p>Next, you will compute the <b><i>covariance matrix</i></b> for the dataset</p>

```
covariance = cov(data_normal.T)
print("The covariance matrix of the normalized data is the following: ")
print(covariance)
```

<p>The next step is to compute the <b><i>eigenvalues</i></b> and <b><i>eigenvectors</i></b> of the dataset.</p>

```
values, vectors = eig(covariance)
print("The eigenvalues of the normalized data are the following: ")
print(values)
```

```
#New lines of code to sort the eigenvalues in descending order, as requested in the question
sorted_indices = np.argsort(values)[::-1]
values = values[sorted_indices]

print("The sorted eigenvalues of the normalized data are the following: ")
print(values)
```

<p>Next, you will select the most important <b><i>eigenvalues</i></b> and adjust the corresponding <b><i>eigenvectors</i></b> of the dataset accordingly.</p>

```
new_values = values[0:3]
print("The most important eigenvalues are the following: ")
print(new_values)
new_vectors = vectors[0:3]
print("The most important eigenvectors are the following: ")
print(new_vectors)
```

<p>Now, you will apply the new <b><i>eigenvectors</i></b> to the old dataset in order to obtain the new, reduced-size dataset.</p>

```
new_data = new_vectors.dot(data_normal.T)
```

The PCA algorithm using the covariance method begins by calculating the mean of each feature ($\bar{x}_j = \frac{1}{n} \sum x_{i,j}$) and subtracting these means from the original dataset $X$ to create a zero-centered matrix ($X_{norm} = X - \bar{x}$). Next, it computes the covariance matrix of the normalized data ($cov(X,Y) = \frac{1}{n-1} \sum (X_{i}- \bar{x})(Y_{i}- \bar{y}$) (we will use $cov$ as $C$) to measure how the features vary together. The algorithm then performs eigendecomposition on this matrix to determine its eigenvalues $\lambda$ and eigenvectors $\mathbf{v}$ ($C\mathbf{v} = \lambda\mathbf{v}$). Instead of sorting, the code simply takes the first three eigenvalues and their corresponding eigenvectors (via the slice [0:3]) to form a projection matrix $W$. Finally, the mean-centered dataset is multiplied by this projection matrix ($X_{new}^T = W X_{norm}^T$) to project and transform the data into a new, reduced-dimensional feature space.

If the value is positive, then that indicates that both dimensions increase together. If the value is negative, then as one dimension increases, the other decreases.

Next, you will load the datasets that will be used to train the <b><i>logistic classification</i></b> model. In the first case, you will train the model using the file <i><a href="https://github.com/netmode/Stochastic-Processes-and-Optimization-in-Machine-Learning-Lab/blob/main/lab2/demo3a.csv">demo3a.csv</a></i>.
```
df = pd.read_csv("https://github.com/netmode/Stochastic-Processes-and-Optimization-in-Machine-Learning-Lab/blob/main/lab2/demo3a.csv")
#df = pd.read_csv("foo.csv")
```

<p>Then, you will split the dataset into a <i>training</i> and a <i>test set</i>.</p>

```
# Separate the input features from the target variable
x = df.iloc[:,1:13].values
y = df.iloc[:,0].values
# Split the dataset into train and test set
Xtrain,Xtest,Ytrain,Ytest = train_test_split(x,y, test_size = 0.2)
```

<p>Then, you will train the <b><i>Logistic Classification</i></b> model.</p>

```
classifier = LogisticRegression(max_iter = 1000)
classifier.fit(Xtrain,Ytrain)
```

<p>Finally, you will obtain the predictions of the model you trained on the test set and compute the total number of errors.</p>

```
# Get the predictions on the test set
prediction = classifier.predict(Xtest)

# Calculate the total number of errors on the test set
errors = 0
for index in range(0,len(prediction) - 1):
	if prediction[index] != Ytest[index]:
		errors += 1

print("Total errors on the test dataset")
print(errors)
```
